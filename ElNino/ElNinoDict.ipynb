```python
def smootherGauss(x, n):     coeff = [0.242, 0.517, 0.242]     #coeff = [0.067, 0.242, 0.383, 0.242, 0.067]## these are coefficients corresponding to Gaussian kernel of width 3## taken from the web page homepages.inf.ed.ac.uk/rbf/HIPR2/gsmooth.htm     x_smooth = []     L = len(x)     store = zeros((n,1), float)  ## array of zeroes for each group of      for u in range(0, L-n):      ## used to produce smoothed value          for v in range(0, n):   ##  add the values multiplied by gaussian coefficients               store[v] = x[u+v]          av = 0          for i in range(0, len(coeff)):               av += float(coeff[i]) * float(store[i])          x_smooth.append(av)     for u in range(L-n,L):  ## this takes care of the "residuals" at the end          for v in range(0, L-u-1):  ## average the values and find the smoothed value               store[v] = x[u+v]          av = 0          for i in range(0, len(coeff)):               av += float(coeff[i]*store[i])          x_smooth.append(av)     return x_smooth                            def plotComparison(raw, smoothed, yrs, label, title, dat1, dat2):    fig = plt.figure(figsize=(10, 10))    plt.xlabel('year')    plt.ylabel(label)    plt.title(title)    plt.plot(yrs, raw, label=dat1, color='red')    plt.plot(yrs, smoothed, label=dat2, color='blue')    plt.legend(loc='lower left')    #plt.show()
```

```python
totElNinoSmooth = smootherGauss(totalElNino, 3) #avgElNinoSmooth = smootherGauss(avgTemp, 5) ## data smoothed over interval of 5 data pointsplotComparison(totalElNino, totElNinoSmooth, yearsElNino, 'El Nino values', 'comparison of total raw with gaussian smoothed data', 'raw data', 'smoothed data')plt.savefig("Gaussian-smoothed.png")
```

```python
def fit(time, a1, a2, T1, T2, phi1, phi2, dphi):    fitFn = []    for i in range(0, int(len(time)/3)):        temp = a1*np.sin(time[i]*1/T1 + phi1) + a2*np.cos(time[i]*1/T2 + phi2)        fitFn.append(temp)    for i in range(int(len(time)/3), len(time)):        temp = a1*np.sin(time[i]*1/T1 + phi1) + a2*np.cos(time[i]*1/T2 + phi2 + dphi)        fitFn.append(temp)              return fitFn
```
