```python
import numpy as npimport matplotlib.pyplot as plt## read data into arrays of floats by yeardef readData(textData):        infile = open(textData,"r") ## read the data file        content = infile.read()        infile.close()        lines = content.split('\n') ## split up lines        data = []        for each in lines:  ## split each line into separate strings                if len(each) < 1:                        continue          ## get rid of the empty list                                           ##at the end                row = each.split()                tot = 0                avg = 0                for i in range(2, len(row)):                    tot += float(row[i])                tot = round(tot, 2)                avg = round(tot/12, 2)                temp = {'lon':float(row[0]), 'lat':float(row[1]), 'jan':float(row[2]), 'feb':float(row[3]), 'mar':float(row[4]), 'apr':float(row[5]), 'may':float(row[6]), 'jun':float(row[7]), 'jul':float(row[8]), 'aug':float(row[9]), 'sep':float(row[10]), 'oct':float(row[11]), 'nov':float(row[12]), 'dec':float(row[13]), 'avg':avg, 'tot':tot}                data.append(temp) ##for each year, longitude, lattitude, january - december, total and average         return data   def precipDuringYear(year):  ## read data during the year of interest        temp = readData("PrecipData/precip."+str(year))        return temp## list of precipitation data for each yeardef temperatureDuringYear(year):  ## read data during the year of interest        temp = readData("TempData/air_temp."+str(year))        return temp## list of temperature data for each year## reduce data based on time which is either total or averagedef dataReduce(data, time, min, max):     lat_vals = []    lon_vals = []    data_vals = []    for i in range(0, len(data)):         temp1 = data[i]['lon'] ## lattitude values        temp2 = data[i]['lat'] ## longitude values        temp3 = data[i][time]        if temp3 <= min or temp3 >=max:            data_vals.append({'lon':temp1, 'lat':temp2, time:temp3})    return data_valsdef dataPlot(data, time, title, min, max):    lat_vals = []    lon_vals = []    data_vals = []    for i in range(0, len(data)):         temp1 = data[i]['lon'] ## longitude values        temp2 = data[i]['lat'] ## lattitude values        temp3 = data[i][time] ## average values        lon_vals.append(temp1)        lat_vals.append(temp2)        data_vals.append(temp3)    fig = plt.figure(figsize=(12, 6))    plt.xlabel('longitude')    plt.ylabel('lattitude')    plt.title(title)    plt.scatter(lon_vals, lat_vals, c=data_vals, edgecolors='none', vmin = min, vmax = max)    plt.colorbar()    plt.show()
```
